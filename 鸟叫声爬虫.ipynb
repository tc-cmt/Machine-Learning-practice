```python
'''
文档说明：
- 功能：
    爬取网页中的mp3或wav格式的链接并命名放入文件夹，一种鸟为一个文件夹
- 参数：
    url中除了pg（页数）的值，其他网址内容都需要输入
- 未解决bug:
    网页内容不止有有名字的，无名字的，还有一种形式为Soundscape，爬取形式和其他的不同，当网页中存在这一项时，会使得声音链接和鸟叫声名字错乱匹配。已使用正则匹配，xpath,soup.find().findAll()来匹配td的内容（具体见试验获取td标签中的Soundscape（未成功）中的代码）
'''

from urllib.request import urlretrieve
url = "https://www.xeno-canto.org/explore?dir=0&order=xc&pg="

for i in range(1,100,1):
    time.sleep(random.randint(5, 6))
    URL = url+str(i)
    html = requests.get(URL).text
    soup = BeautifulSoup(html, 'lxml')
    voice_links = soup.find_all("div", {"data-xc-filepath": re.compile('.*?\.mp3|\.wav')})#鸟叫声
    voice_name = soup.find_all("span",{"class":"common-name"})#鸟的种类
    voice_name2 = soup.find_all("span",{"title":"ID under discussion"}) 
   
    print(i,len(voice_links))#鸟叫声个数
    os.makedirs('./voiceofBird/', exist_ok=True)#创建voice文件夹
    for ul,n in zip(voice_links,voice_name):
        time.sleep(random.randint(1, 2))
        print(ul['data-xc-filepath'])
        VOICE_URL = ul['data-xc-filepath']
        VOICE_URL = "https:"+VOICE_URL
        r = requests.get(VOICE_URL)
        
        name = n.get_text()
        os.makedirs('./voiceofBird/'+name+'/', exist_ok=True)#创建子文件夹
        num = re.match("^XC\d+",VOICE_URL.split("/")[-1]).group()
        print(num)
        bird_name = name+'-'+num+'.mp3'
        with open('./voiceofBird/%s/%s' % (name,bird_name), 'wb') as f:
            f.write(r.content)
        print('Saved %s' % name)



        
```
